```python
n = int(input("Enter a number: "))

result = (n > 0) - (n < 0)

print(["Negative", "Zero", "Positive"][result + 1])
```

```python
year = int(input("Enter a year: "))

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print("Leap Year")
else:
    print("Not a Leap Year")
```

```python
##### num1 = float(input("Enter first number: "))
op = input("Enter operator (+, -, *, /): ")
num2 = float(input("Enter second number: "))

result = {
    '+': num1 + num2,
    '-': num1 - num2,
    '*': num1 * num2,
    '/': num1 / num2
}

print("Result =", result[op])
```

```python
ch = input("Enter a character: ")

result = ["Consonant", "Vowel"][ch in 'aeiouAEIOU']
print(result)
```

```python
marks = int(input("Enter marks: "))

if marks >= 90:
    print("Grade: A")
elif marks >= 75:
    print("Grade: B")
elif marks >= 50:
    print("Grade: C")
else:
    print("Grade: D")
```
